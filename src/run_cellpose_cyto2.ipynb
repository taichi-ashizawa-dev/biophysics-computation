```python
# --- 手順1: 必要なライブラリのインストール ---
!pip install "cellpose[gui]" --quiet

import os
import numpy as np
import matplotlib.pyplot as plt
from cellpose.models import Cellpose # Cellposeクラスをcellpose.modelsから直接インポート
from cellpose import io, utils # その他のモジュールは通常通りインポート

# --- 手順2: LIVECell学習済みモデルの読み込み ---
# Cellposeクラスはmodelsモジュール内にあります
model = Cellpose(gpu=True, model_type='cyto2') # 直接インポートしたので models. は不要
print("✅ LIVECell学習済みモデル(cyto2)の読み込みが完了しました。")

# --- 手順3: 画像の読み込みと推論 ---

# ★ここにあなたの画像のパスを貼り付けてください★
# 例: img_path = '/content/20260126_experiment_01.tif'
# 何も変更しなければ、強制的に新しいサンプル画像をダウンロードします
img_path = 'sample_image_v2.png'

# ユーザーの画像指定がない、またはファイルがない場合はサンプルをダウンロード
if img_path == 'sample_image_v2.png' and not os.path.exists(img_path):
    # img03.png (細胞質が大きく写っているサンプル) を取得してみます
    io.imsave(img_path, utils.download_url_to_file('https://www.cellpose.org/static/images/img03.png', img_path, progress=True))

print(f"📂 画像を読み込んでいます: {img_path}")
image = io.imread(img_path)

# 推論実行
# channels=[0,0] はグレースケール（位相差画像）用の設定です
# diameter=None とすると、AIが細胞のサイズを自動推定します
masks, flows, styles, diams = model.eval(image, diameter=None, channels=[0,0])

# 結果の表示
fig = plt.figure(figsize=(12,5))
ax = fig.add_subplot(1,2,1)
ax.imshow(image, cmap='gray')
ax.set_title('Input Image')
ax = fig.add_subplot(1,2,2)
ax.imshow(masks)
ax.set_title('Segmentation Result (cyto2)')
plt.show()
```
